# B. Visualizações simples <br>

## Bi. Visualização da distribuição entre as classes
<a id="dist_class"></a>

> É muito importante sabermos como as classes que serão utilizadas para a construção de um modelo estão distribuidas, conjunto de dados com classes desbalançeadas requer algumas mudanças no processo de modelagem, como por exemplo a estratificação das classes. Para que possamos fazer uma boa visualização de como está a distribuição entre as classes, abaixo econtra-se um exemplo simples e bonito de como fazer uma boa visualização da distribuição das classes de um conjunto de dados.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Criando Dados**

---

<b>
    
0 -> Cachorro

1 -> Porco

2 -> Burro
    
</b>

```python
# Fixando um seed
np.random.seed(20)

df = pd.Series(np.random.randint(0, 3, 1000))
```

**Criando Visualização**

```python
# Tamanho e estilo do gráfico
plt.style.use('Solarize_Light2')
fig = plt.figure(figsize = (5, 5))
plt.style.use('classic')

# Definindo o gráfico I and II
eixo1 = fig.add_axes([0, 0, 1, 1]) # Graph - Bars
eixo2 = fig.add_axes([1.1, 0, 1, 1]) # Graph - Pie

# ---= Criando os Gráficos =---

# Gráfico I - Bars
#########################################################
cores = ["aqua", "orange", "lime"]
indice = ['Cachorro', 'Porco', 'Burro']
valores = df.value_counts().values

# Gráfico
eixo1.bar(indice, valores, lw = 1.5, edgecolor='k', color = cores)

# Labels
eixo1.set_ylabel('Frequência', fontsize = 14)
eixo1.set_title('Distribuição entre as classes', fontsize = 16, pad=10)

# Anotações
eixo1.annotate(f"{df.value_counts().values[0]}", xy=(0, 365), fontsize = 15, color = cores[0])
eixo1.annotate(f"{df.value_counts().values[1]}", xy=(1, 330), fontsize = 15, color = cores[1])
eixo1.annotate(f"{df.value_counts().values[2]}", xy=(2, 317), fontsize = 15, color = cores[2])

# Gráfico II - Pies
#########################################################

# Gráfico
eixo2.pie(df.value_counts(), colors=cores, autopct='%.2f%%', textprops={'size' : 13}, labels=indice,
         explode=(0.03, 0, 0), wedgeprops=dict(width=.6), shadow = True)

# Título
eixo2.set_title('Distribuição percentual entre as classes', fontsize = 15, pad=10);
```
